```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

Begin with two variables `(fig, ax)`. fig is assigned the `plt.figure()` method with the argument `figsize=(8,8)` that establishes the overall size of the chart. `ax` is assigned `fig.add_suplot()` with arguments that also help construct the size of the chart. In the end, we add `plt.show()`, which displays the chart, and `plt.grid()`, which displays gridlines.

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)
plt.grid()
plt.show()
```

Now we have a blank canvass or graph from which we can begin drawing our emoji. Let’s start with the face shape first — the eyes and mouth will come later.
To create the face, we’ll call `ax.scatter()` to create what would be a scatter plot, but with just one point. From inside the `scatter()` method, we have the arguments:

To create the face, we’ll call `ax.scatter()` to create what would be a scatter plot, but with just one point. From inside the `scatter()` method, we have the arguments:

- `[.5],[.5]`; the point’s coordinates
- `c` and `edgecolor = yellow` and `gold` respectively to color in the face and edge.
- `linewidth`; affects the thickness of the outer edge
- `s`; affects the overall size of the face.

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)

ax.scatter([.5],[.5], c="yellow", edgecolor = "gold", linewidth = 15, s = 130000)

plt.grid()
plt.show()
```

We’ve now created the sun, but that’s not why we’re here. Let’s continue by adding in the first eye.
Add in another `ax.scatter()` method below the first one we created for the eye. But this time, we’ll be changing the coordinates to `[.37],[.60]` which can be adjusted if you wish to do so. The color(`c`) will be darkgoldenrod with no need for an edge color. The size(`s`) will be changed to 1700.

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)
ax.scatter([.5],[.5], c="yellow", edgecolor = "gold", linewidth = 15, s = 130000)


ax.scatter([.37],[.60], c = "darkgoldenrod", s = 1700)


plt.grid()
plt.show()
```

Instead of the sun, now it kind of looks like a lemon. However, we’re not interested in drawing lemons, and so let’s keep going by adding in the smile.

Now, this time we won’t be using the `ax.scatter()` method. Instead, we’ll be implementing something a little bit more complicated, but not that complicated. First, we’re going to be making two new variables: `X` and `Y`. Then, we’ll plot the `X` and 'Y' with the 'ax.plot()' method.

For `X = np.linspace(.37, .63, 100)`, we’re finally using a numpy method: `np.linspace()` with three arguments. The first two arguments represent the start and endpoints on the x-axis (.37, .63). The third represents the number of points to graph (in this case, 100 points).

For `Y = 4 * (X-.5)**2 + 0.3`, we have an equation with four numbers that we can modify. The first number (`4`) affects the curve/angle of the smile. If it was negative, it would’ve been a frown. The second number (`.5`) alters the rotation of the smile just slightly. The third number (`2`, the exponent) affects how straight the smile is. The fourth number (`0.3`) alters the position of the smile on the y-axis.

Finally, with `ax.plot(X, Y, c=’darkgoldenrod’, linewidth=20)`, we plot our (`X, Y`) variables we created with the color(`c`), the same as the eye (darkgoldenrod), and the linewidth equal to `20`, which modifies the smile thickness.


```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)
ax.scatter([.5],[.5], c="yellow", edgecolor = "gold", linewidth = 15, s = 130000)
ax.scatter([.37],[.60], c = "darkgoldenrod", s = 1700)

X = np.linspace(.37, .63, 100)
Y = 4 * (X - .5)**2 + 0.3
ax.plot(X, Y, c = "darkgoldenrod", linewidth = 20)

plt.grid()
plt.show()
```

Moving on to the eyebrows and the winking eye, we must repeat the process that we did for the smile. We’re essentially plotting three other lines with the same method we just used. For brevity’s sake, feel free to replicate the code below.
As you can see, it’s exactly in the same format as the smile’s code, but with different values and assigned to new variable names. The values currently used appear to fit best, but feel free to adjust it to whatever value you like.

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)
ax.scatter([.5],[.5], c="yellow", edgecolor = "gold", linewidth = 15, s = 130000)
ax.scatter([.37],[.60], c = "darkgoldenrod", s = 1700)
X = np.linspace(.37, .63, 100)
Y = 4 * (X - .5)**2 + 0.3
ax.plot(X, Y, c = "darkgoldenrod", linewidth = 20)



a = np.linspace(.6, .7, 100)
b = -1.75 * (a - .7)**2 + 0.6

c = np.linspace(.63, .75, 100)
d = - 1.25 * (a - .6) ** 2 + 0.7

e = np.linspace(.25, .35, 100)
f = - 2 * (a - .8) ** 2 + 0.77

ax.plot(a,b, c = "darkgoldenrod", linewidth = 11)
ax.plot(c,d, c = "darkgoldenrod", linewidth = 9)
ax.plot(e,f, c = "darkgoldenrod", linewidth = 9)





plt.grid()
plt.show()
```

We did it — we finally have our emoji! But, what about the grid lines and plot axes? Well, we can make them vanish with just a few more lines of code. Don’t worry, these are the last bits of code that we have to write.

We remove the lines from the graph by using the `ax.spines.set_visible()` method by selecting correct sides in the ax.spines brackets and passing in Falsein the `set_visible()` method. Finally, we use `ax.set_xticks()` and `ax.set_yticks()` to an empty list to erase the numbers on the x and y axes.

```python
fig = plt.figure(figsize= (8,8))
ax = fig.add_subplot(1,1,1, aspect = 1)
ax.scatter([.5],[.5], c="yellow", edgecolor = "gold", linewidth = 15, s = 130000)
ax.scatter([.37],[.60], c = "darkgoldenrod", s = 1700)
X = np.linspace(.37, .63, 100)
Y = 4 * (X - .5)**2 + 0.3
ax.plot(X, Y, c = "darkgoldenrod", linewidth = 20)
a = np.linspace(.6, .7, 100)
b = -1.75 * (a - .7)**2 + 0.6
c = np.linspace(.63, .75, 100)
d = - 1.25 * (a - .6) ** 2 + 0.7
e = np.linspace(.25, .35, 100)
f = - 2 * (a - .8) ** 2 + 0.77
ax.plot(a,b, c = "darkgoldenrod", linewidth = 11)
ax.plot(c,d, c = "darkgoldenrod", linewidth = 9)
ax.plot(e,f, c = "darkgoldenrod", linewidth = 9)



ax.spines['top'].set_visible('False')
ax.spines['bottom'].set_visible('False')
ax.spines['right'].set_visible('False')
ax.spines['left'].set_visible('False')

ax.set_xticks([])
ax.set_yticks([])


plt.grid()
plt.show()
```
